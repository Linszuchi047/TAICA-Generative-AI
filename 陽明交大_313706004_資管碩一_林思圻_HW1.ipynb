```python
# prompt: Áï´Âá∫ÂÉèËù¥Ëù∂ÁöÑÂáΩÊï∏ÂúñÂΩ¢

import numpy as np
import matplotlib.pyplot as plt

# Define the function for a butterfly-like shape
def butterfly_function(t):
    x = np.sin(t) * (np.exp(np.cos(t)) - 2 * np.cos(4 * t) - np.sin(t / 12) ** 5)
    y = np.cos(t) * (np.exp(np.cos(t)) - 2 * np.cos(4 * t) - np.sin(t / 12) ** 5)
    return x, y

# Generate a range of t values
t = np.linspace(0, 12 * np.pi, 500)

# Calculate x and y values using the butterfly function
x, y = butterfly_function(t)

# Create the plot
plt.figure(figsize=(8, 8))
plt.plot(x, y, color='blue')

# Customize the plot (optional)
plt.title("Butterfly-like Function")
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')  # Ensure the x and y axes have the same scale
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box') # Maintain aspect ratio
plt.show()
```
